```python
df = pd.read_csv('https://drive.google.com/uc?id=1a0VKzKV6NSmbKMRHSfxxxFLZe8yJVbtO')
df
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import MiniBatchKMeans

df = df.dropna()

# Векторизация текста с использованием TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['text'])

```

```python
# Кластеризация текстов на основе векторов
n_clusters = 39
clustering = MiniBatchKMeans(n_clusters=n_clusters, verbose=1, batch_size=100)
clusters = clustering.fit_predict(X)
```

```python
# Добавление колонки с кластерами в датафрейм
df['cluster'] = clusters
```

```python
# Сохранение результатов
df.to_csv('clustered_data.csv', index=False)
```

```python
# Функция для определения схожести текстов
def text_similarity(text1, text2):
    return fuzz.ratio(text1, text2)

# Обновленная функция для удаления дубликатов с учетом семантической близости
def remove_duplicates_with_semantics(cluster_id):
    cluster_df = df[df['cluster'] == cluster_id]
    cluster_texts = cluster_df['text'].tolist()
    unique_texts = [cluster_texts[0]]  # Начинаем с первого текста

    for text in cluster_texts:
        # Проверяем схожесть с остальными текстами
        is_duplicate = False
        for unique_text in unique_texts:
            similarity = text_similarity(text, unique_text)
            if similarity > 90:  # Порог схожести (в данном случае 90%)
                is_duplicate = True
                break
        if not is_duplicate:
            unique_texts.append(text)

    cluster_df = cluster_df.iloc[:len(unique_texts)]
    return cluster_df
```

```python
result_df = df.drop_duplicates(subset='cluster')
```

```python
unique_dfs = []
unique_clusters = df['cluster'].unique()

for cluster_id in unique_clusters:
    unique_df = remove_duplicates_with_semantics(cluster_id)
if not unique_df.empty:
  unique_dfs.append(unique_df)
```

```python
result_df = pd.concat(unique_dfs)
result_df.to_excel("результаты_классификации_и_удаления_дубликатов_на_одном_листе.xlsx", index=False)
```
